2. Hitung vektor 3 dimensi

```python
vect_b = np.array([10, 20, 11])
print(vect_b)
vect_c = np.array([5, 12, 9])
print(vect_c)
```

```python
print(vect_a + vect_b)
print(vect_b - vect_c)
print(vect_a @ vect_c)
print(vect_a * vect_b)
print(np.linalg.norm(vect_a))
```

```python
#Sudut antara vektor A dan vektor B
cos_theta = (vect_a @ vect_b) / (np.linalg.norm(vect_a) * np.linalg.norm(vect_b))
print(cos_theta)
```

```python
radian = np.arccos(cos_theta) #Mengubah Cos Theta > Radian
print(radian)
degree = np.degrees(radian) #Mengubah Radian > Sudut
print(degree)
```

#3. Buat plot dari vektor kedalam bidang 2D

```python
U = np.array([2, 5])
V = np.array([3, 1])

import matplotlib.pyplot as plt
```

```python
plt.arrow(0, 0, U[0], U[1], color="green", head_width=0.2)
plt.arrow(0, 0, V[0], V[1], color="black", head_width=0.2)
plt.axis([0, 7, 0, 7])
plt.grid()
plt.show()
```

```python
uv = np.array([[2, 5], [3, 1]])

plt.scatter(uv[:, 0], uv[:, 1], color="red")
plt.axis([0, 6, 0, 6])
plt.grid()
plt.show()
```
