V. Data Exploration

Untuk memudahkan pembacaan pada tanggal, maka dilakukanlah pengelompokkan data per-quarter selama 3 tahun berjalan

```python
selected_column = gc1.iloc[:, 4:]
```

```python
first_date = selected_column.iloc[:,0:1]
first_date
```

```python
last_date = selected_column.iloc[:, -1:]
last_date
```

```python
group_date = pd.concat([first_date, last_date], axis=1)
group_date
```

```python
column_headers = list(group_date.columns)
column_headers
```

pandas melakukan pengelompokkan otomatis selama 3 bulan, dengan maksud nantinya, jika ada penambahan data untuk beberapa bulan kedepan, maka pembacaan program tetap dilakukan secara otomatis, tidak perlu merubah command dari awal lagi.

```python
transform_date = []
for date in column_headers:
    split_date = date.split("_")
    split_date.pop(0)
    split_date[-1] = "20"+str(split_date[-1])
    split_date = [int(i) for i in split_date]
    transform_date.append(datetime(split_date[-1], split_date[0], split_date[1]))
    
quarter = pd.date_range(pd.to_datetime(transform_date[0].date()), pd.to_datetime(transform_date[1].date()) + pd.offsets.QuarterBegin(1), freq='Q').strftime('_%-m_%-d_%y').tolist()
quarter
```

```python
get_quarter_column = gc1[gc1.columns.intersection(quarter)]
get_quarter_column
```

```python
selected_column = gc1.iloc[:, 0:]
```

```python
country = selected_column.iloc[:,0:1]
country
```

```python
group_deaths = pd.concat([country, get_quarter_column], axis=1)
group_deaths
```

Agar dapat dibaca dengan baik oleh python, maka "_" harus dihapuskan, sehingga diperlukannya pergantian nama disini.

```python
group_deaths.rename(columns={u'_3_31_20' : u'31-Mar-20'}, inplace=True)
group_deaths.rename(columns={u'_6_30_20' : u'30-Jun-20'}, inplace=True)
group_deaths.rename(columns={u'_9_30_20' : u'30-Sep-20'}, inplace=True)
group_deaths.rename(columns={u'_12_31_20' : u'31-Dec-20'}, inplace=True)
group_deaths.rename(columns={u'_3_31_21' : u'31-Mar-21'}, inplace=True)
group_deaths.rename(columns={u'_6_30_21' : u'30-Jun-21'}, inplace=True)
group_deaths.rename(columns={u'_9_30_21' : u'30-Sep-21'}, inplace=True)
group_deaths.rename(columns={u'_12_31_21' : u'31-Dec-21'}, inplace=True)
group_deaths.rename(columns={u'_3_31_22' : u'31-Mar-22'}, inplace=True)
group_deaths.rename(columns={u'_6_30_22' : u'30-Jun-22'}, inplace=True)
group_deaths.rename(columns={u'_9_30_22' : u'30-Sep-22'}, inplace=True)
```
