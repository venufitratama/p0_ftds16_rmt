```python
import sympy as sy
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = sy.Symbol ('x')
y = sy.Symbol ('y')
```

```python
y = 1.25 - 0.75*x
y
```

```python
x = np.linspace(-50, 50)
y = 1.25 - 0.75*x

gradient_x = np.gradient(y)
gradient_x
```

```python
fig, ax = plt.subplots(ncols=2, figsize=(21,5))

ax[0].plot(x, y)
ax[0].set_title('Plot of `x`')

ax[1].plot(gradient_x, y)
ax[1].set_title('Plot of `result_gradient_x`')
plt.xlim(-2, 2)
```

```python
y = np.linspace(-50, 50)

x = (5/3) - (4/3) * x

gradient_y = np.gradient(x)
gradient_y
```

```python
fig, ax = plt.subplots(ncols=2, figsize=(21,5))

ax[0].plot(x, y)
ax[0].set_title('Plot of `y`')

ax[1].plot(x, gradient_y)
ax[1].set_title('Plot of `result_gradient_y`')
plt.ylim(-3, 3)
```
